```python
from prettytable import PrettyTable

users = []
balances = {}
transactions = []


n = int(input("Enter number of users: "))
for i in range(n):
    name = input(f"Enter user {i+1} name: ")
    users.append(name)
    balances[name] = 0

def add_expense():
    payer = input("Enter payer name: ")
    amount = float(input("Enter total amount: "))
    description = input("Enter expense description: ")

    split_type = input("Split type (equal/custom): ")

    if split_type == "equal":
        share = amount / len(users)
        for user in users:
            if user != payer:
                balances[user] -= share
                balances[payer] += share

    elif split_type == "custom":
        for user in users:
            share = float(input(f"Enter amount for {user}: "))
            if user != payer:
                balances[user] -= share
                balances[payer] += share

    transactions.append({
        "payer": payer,
        "amount": amount,
        "description": description
    })

    print("Expense added successfully!\n")

def show_balances():
    table = PrettyTable()
    table.field_names = ["User", "Balance"]

    for user in users:
        table.add_row([user, round(balances[user], 2)])

    print(table)

def show_transactions():
    print("\nTransaction History")
    for t in transactions:
        print(f"{t['payer']} paid ₹{t['amount']} for {t['description']}")

while True:
    print("\n1. Add Expense")
    print("2. Show Balances")
    print("3. Show Transactions")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        add_expense()
    elif choice == 2:
        show_balances()
    elif choice == 3:
        show_transactions()
    elif choice == 4:
        break
    else:
        print("Invalid choice")
```
